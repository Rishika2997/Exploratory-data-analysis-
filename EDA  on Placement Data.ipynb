# using relplot when two variable are numeric
- relplot is combination of scatter plot and line plot by defualt

```python
sns.relplot("etest_p", "salary", data=data, hue='degree_t', style='degree_t');
```

```python
sns.relplot("mba_p", "salary", data=data, hue="specialisation", style="specialisation");
```

```python
sns.relplot("mba_p","salary",data=data,kind="line",col="specialisation",hue="workex");
```

Here, we can see that when the "specialisation" is "Mkt&Fin" then the salary of most of the candidates is high.

```python
plt.bar(data["workex"], data["salary"] , color="green");
plt.xlabel('workex')
plt.ylabel('salary')
plt.title('Salary according to work experience' );
```

Here, we can observe when candidate is having more work experience then salary is higher

```python
sns.relplot(x="degree_t",y="degree_p",data=data,hue="workex",col="status");
```

Candidates who are having a high percentage in degree college have a better cance of placement than the candidates who are having low percentage in degree college.

```python
plt.figure(figsize=(15,7))
plt.subplot(131)
sns.countplot(data['hsc_s'])
plt.title("Higer secondary education")

plt.figure(figsize=(15,7))
plt.subplot(132)
sns.countplot(data['degree_t'])
plt.title("Field of Degree education")

plt.figure(figsize=(15,7))
plt.subplot(133)
sns.countplot(data['specialisation'])
plt.title("Post graduation (MBA)")

```

Here we can observe in most of candidate have choosen commerce and same as well as in degree and in MBA colleges the widely selected specialisation field is "Mkt&Fin"

```python
plt.figure(figsize=(10,7))
plt.subplot(152)
sns.stripplot(x='status', y='hsc_p', data=data)
plt.title("12th percentage")

plt.figure(figsize=(10,7))
plt.subplot(153)
sns.stripplot(x='status', y='degree_p', data=data)
plt.title("degree percentage")

plt.figure(figsize=(10,7))
plt.subplot(154)
sns.stripplot(x='status', y='etest_p', data=data)
plt.title(" Employability test percentage")


plt.figure(figsize=(10,7))
plt.subplot(155)
sns.stripplot(x='status', y='mba_p', data=data)
plt.title("MBA percentage")
```

As we can clearly see that the density of blue dots increase when the percentage increases. So it is very clear that more the percentage ,more the chances of getting placement.

```python
plt.figure(figsize=(7,7))
sns.boxplot(x='gender', y='salary', hue='hsc_s', data=data, palette = ['blue','green', 'red']);
```

Salaries of both the gender are almost same but the science field has a better hand in this as highest salary is from science department.

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(),annot=True);
```
